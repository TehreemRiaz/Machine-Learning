 ### Fit Linear Regression Model (Training data)

 As our linear Model was

)

And using derivatives we transformed our model into 2 simultaneous equations

Then converted it in matrix form

![Model Equations](https://github.com/ahmad-14a/CS-F20-ML/blob/main/Regression-Assignment%202/Linear%20Model.png?raw=true)



NOTE :  Now you are required to compute values wihtout using foor loop on tarining data
- HINT : Use numpy library for this purpose 

```python
#Your code Here
#For loop is not recommended in these cases, 
#use numpy functions to calculate the values of above variables in one line of code.
#Such as x.shape for number of rows, sum() for sum of all elements in x
#np.dot() or multiply suntion for square of x

# Get the number of rows and columns in the training data
rows = trainData.shape[0]
cols = trainData.shape[1]

# Calculate the sum of x_train
sumx = np.sum(x_train)

# Construct matrix A using the number of rows and the sum of x_train
A = np.array([[rows, sumx],
             [sumx, np.sum(np.dot(x_train, x_train))]])
print('Matrix A\n\n', A)
# Construct matrix B using the sum of y_train and the dot product of x_train and y_train
B = np.array([[np.sum(y_train)],
              [np.sum(np.dot(x_train, y_train))]])
print('\nMatrix B\n\n', B)
```

Expected output

![Linear Model MAtrixes](https://github.com/ahmad-14a/CS-F20-ML/blob/main/Regression-Assignment%202/Matrixs%20Linear.png?raw=true)


#### Compute Valus of Both Î˜`s


```python
#Your code Here
#Calculate the values of 0_node and 0_1 using Matrix Multiplication
#Hint: Use X = A^-1 * B to calculate the values of 0_node and 0_1

# Calculate the inverse of matrix A
Ainv = np.linalg.inv(A)

# Multiply the inverse of A with B to get the coefficients for the linear model
x = np.dot(Ainv, B)
print('Matrix X for linear model\n\n', x)
```

Expected Output

### Run Predictions (Testing data)

```python
# Your code Here
# In order to calculate the pridicted values of y
# Use the formula y = 0_node + 0_1 * x
# _____DO NOT USE FOR LOOP_____
# instead Convert the column of test_x into numpy a 2 dimensional matrix
# then use matrix multiplication to calculate the predicted values of y


# predTEST = x[0][0] + np.dot(x[1][0], x_test)
# print(predTEST)
# print(y_test)

coef_matrix = np.array([[-0.55251074], [0.8736061]])

# Add a column of ones to x_test and store it in test_x_matrix so that we multiply it
# like this [theta_1, theta_node] * [x_test, ones], Y = theta_1 * x_test + theta_node * ones
test_x_matrix = np.column_stack((x_test, np.ones(len(x_test))))

# Calculate the predicted y values for the test data
predtest = np.dot(test_x_matrix, coef_matrix)
```

### Mean Square Error

![Cost Funtion](https://github.com/ahmad-14a/CS-F20-ML/blob/main/Regression-Assignment%202/Cost%20funtion1.png?raw=true
)

```python
#Your code Here
#Using cost function calculate the cost of the model

# Get the number of rows in the test data
rowstest = testData.shape[0]

# Calculate the cost as the sum of the squared differences between the predicted and actual y values
cost = np.square(np.subtract(predtest.flatten(), y_test))

# Calculate the mean square error
err = np.sum(cost) / (rowstest)
print('Mean Square Error for LINEAR MODEL is:    ', err)
```

```python
##Your code Here
coef_matrix = np.array([[-0.55251074], [0.8736061]])

# Add a column of ones to x_train and store it in train_x_matrix
train_x_matrix = np.column_stack((x_train, np.ones(len(x_train))))

# Calculate the predicted y values for the training data
predTRAIN = np.dot(train_x_matrix, coef_matrix)

# Plot the training data and the fitted linear model
plt.scatter(x_train, y_train, color='purple', label='Train Data')
plt.plot(x_train, predTRAIN, color='red', label='Linear Model')
plt.xlabel('X')
plt.ylabel('R')
plt.title('Linear Model for the train data')
plt.legend()
plt.show()

# Plot the test data and the fitted linear model
plt.scatter(x_test, y_test, color='purple', label='Train Data')
plt.plot(x_test, predtest, color='red', label='Linear Model')
plt.xlabel('X')
plt.ylabel('R')
plt.title('Linear Model for the train data')
plt.legend()
plt.show()
```
